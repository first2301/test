# prerpro train data

```python
train_path = glob.glob("../data/Small_Business_Customer_Order_QnA/Training/라벨링데이터_train/*.csv")
```

```python
train_path
```

```python
df0, df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13 = [pd.read_csv(path) for path in train_path]
```

```python
df0
```

```python
def process_df(df):
    df_processed = df[["QA여부", "발화문"]].copy()
    df_processed["QA여부"].replace({"q": "user", "a": "system"}, inplace=True)
    df_processed.dropna(inplace=True)
    df_processed.reset_index(drop=True, inplace=True)
    return df_processed

[df0, df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13] = [
    process_df(df) for df in [df0, df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13]
]
```

```python
csv_name = []
for name in train_path:
    csv_name.append(name.split("/")[-1].split(".")[0])  # "건강_train"
```

```python
csv_name
```

```python
all_conversations = []

for df in [df0, df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13]:
    conversations = []
    current_dialogue = []
    
    for idx, row in df.iterrows():
        role = row["QA여부"]
        content = row["발화문"].strip()

        if role not in ["user", "system"]:
            continue

        current_dialogue.append({
            "role": role,
            "content": content
        })

        if len(current_dialogue) >= 2 and \
           current_dialogue[-2]["role"] == "user" and current_dialogue[-1]["role"] == "system":
            conversations.append({"messages": current_dialogue.copy()})
            current_dialogue = []
    
    all_conversations.extend(conversations)

for name in csv_name:
    with open(f"../data/Small_Business_Customer_Order_QnA/Training/labeling_data_train_json/{name}.jsonl", "w", encoding="utf-8") as f:
        for convo in all_conversations:
            f.write(json.dumps(convo, ensure_ascii=False) + "\n")

print(f"✅ 변환 완료! 총 {len(all_conversations)}개 대화가 생성되었습니다.")
```
